```python
import joblib

# Define the path to your saved model file
model_filename = r'C:\Users\jaiku\PycharmProjects\EV_Predictive_Maintenance\models\optimized_soh_xgb_model.joblib'

# Load the model from the file
try:
    loaded_model = joblib.load(model_filename)

    # Access and print the feature names
    if hasattr(loaded_model, 'feature_names_in_'):
        feature_names = loaded_model.feature_names_in_
        print("✅ Feature names found:")
        for i, feature in enumerate(feature_names):
            print(f"{i + 1}: {feature}")
    else:
        print("⚠️ Could not find the 'feature_names_in_' attribute.")
        print("The model may not have been trained on a DataFrame with column names.")

except FileNotFoundError:
    print(f"❌ Error: The file '{model_filename}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```
